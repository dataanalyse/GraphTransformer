```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

def box(ax, x, y, w, h, text):
    p = FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.02", linewidth=1.5)
    ax.add_patch(p)
    ax.text(x + w/2, y + h/2, text, ha="center", va="center", fontsize=10)

fig, ax = plt.subplots(figsize=(10, 2.6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 3)
ax.axis("off")

# Pipeline: X(t) + Graph -> Graph Layer -> Predictor -> Y(t+1)
box(ax, 0.3, 1.8, 2.2, 0.8, "Node features X(t)\n[health, exposure, TTR]")
box(ax, 0.3, 0.4, 2.2, 0.8, "Graph structure\n(edge_index)")

box(ax, 3.1, 1.1, 2.2, 0.8, "Graph Layer\n(GCN / GAT)")
box(ax, 5.9, 1.1, 2.2, 0.8, "Classifier\n(sigmoid)")
box(ax, 8.7, 1.1, 1.0, 0.8, "Y(t+1)\nhealth")

# Arrows
ax.annotate("", xy=(3.1, 1.5), xytext=(2.5, 2.2), arrowprops=dict(arrowstyle="->", lw=1.5))
ax.annotate("", xy=(3.1, 1.5), xytext=(2.5, 0.8), arrowprops=dict(arrowstyle="->", lw=1.5))
ax.annotate("", xy=(5.9, 1.5), xytext=(5.3, 1.5), arrowprops=dict(arrowstyle="->", lw=1.5))
ax.annotate("", xy=(8.7, 1.5), xytext=(8.1, 1.5), arrowprops=dict(arrowstyle="->", lw=1.5))

plt.tight_layout()
plt.savefig("model_block_diagram.png", dpi=300)
print("Wrote model_block_diagram.png")
```
